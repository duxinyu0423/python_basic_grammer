打开文件 <br>
- open('文件名字', '打开模式', '编码方式')

```python
f = open('./read.txt', 'r', encoding = 'utf-8')

# print(f.read(20))   #读取前 20 个字符
# print(f.readline())  #读取一行

lines = f.readlines()
print(lines)
for line in lines:
    print(line)

f.close()
```

line = f.readline() <br>
while line != '': <br>
    print(line) <br>
    line = f.readline() <br>



with打开自动关闭

```python
with open('./read.txt') as f:
    print(f.read())
```
